This competition presents a chance to benchmark your sentiment-analysis ideas on the Rotten Tomatoes dataset. You are asked to label phrases on a scale of five values: negative, somewhat negative, neutral, somewhat positive, positive. Obstacles like sentence negation, sarcasm, terseness, language ambiguity, and many others make this task very challenging.

Effectively was 52 words, this means if we would Tokenize by word the max_length should be 52, however as transformers consider sub-words tokenization such number could be increased depending on the words being used which can increase such length to 60 or even more, thus we have to take that into account when modeling as it could cause our model to take significatively a long time to train, therefore we have to find a trade-off between training time and performance.

```python
dfff=pd.DataFrame(df['Phrase'].str.split().map(lambda x: len(x))>=20)
print('Number of sentences which contain more than 20 words: ', dfff.loc[dfff['Phrase']==True].shape[0])
print(' ')
dfff=pd.DataFrame(df['Phrase'].str.split().map(lambda x: len(x))>=30)
print('Number of sentences which contain more than 30 words: ', dfff.loc[dfff['Phrase']==True].shape[0])
print(' ')
dfff=pd.DataFrame(df['Phrase'].str.split().map(lambda x: len(x))>=40)
print('Number of sentences which contain more than 40 words: ', dfff.loc[dfff['Phrase']==True].shape[0])
print(' ')
dfff=pd.DataFrame(df['Phrase'].str.split().map(lambda x: len(x))>=50)
print('Number of sentences which contain more than 50 words: ', dfff.loc[dfff['Phrase']==True].shape[0])
print(' ')
dfff=pd.DataFrame(df['Phrase'].str.split().map(lambda x: len(x))==52)
print('Number of sentences which contain 52 words: ', dfff.loc[dfff['Phrase']==True].shape[0])
print(' ')
#dfff.loc[dfff['Phrase']==True]
```

```python
from transformers import BartTokenizer, TFBartModel
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.initializers import TruncatedNormal
from tensorflow.keras.utils import to_categorical

# Define max_length
max_length = 45

# Define your model creation function
def create_model(max_length):
    with strategy.scope():
        # Load the BART model and tokenizer
        bart_model = TFBartModel.from_pretrained('facebook/bart-base')
        tokenizer = BartTokenizer.from_pretrained('facebook/bart-base')

        # Get the main layer
        bart = bart_model.layers[0]

        # Build your model input
        input_ids = Input(shape=(max_length,), name='input_ids', dtype='int32')
        inputs = {'input_ids': input_ids}

        # Load the Transformers BART model as a layer in a Keras model
        bart_outputs = bart(inputs)[0]
        bart_model_output = tf.squeeze(bart_outputs[:, -1:, :], axis=1)
        dropout = Dropout(0.1, name='pooled_output')
        pooled_output = dropout(bart_model_output, training=False)

        # Build the model output
        Sentiments = Dense(units=len(data_train.Sentiment_label.value_counts()), kernel_initializer=TruncatedNormal(stddev=0.02), name='Sentiment')(pooled_output)
        outputs = {'Sentiment': Sentiments}

        # Combine it all in a model object
        model = Model(inputs=inputs, outputs=outputs, name='BART_MultiClass')

        # Compile the model
        optimizer = Adam(learning_rate=5e-05, epsilon=1e-08, clipnorm=1.0)
        loss = {'Sentiment': CategoricalCrossentropy(from_logits=True)}
        metrics = {'Sentiment': 'accuracy'}
        model.compile(optimizer=optimizer, loss=loss, metrics=metrics)

        return model

# Print the model summary
model_bart = create_model(max_length)
model_bart.summary()
```

```python
from transformers import AlbertTokenizer, TFAlbertModel
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.utils import to_categorical


def create_model(max_length):
    with strategy.scope():
        # Load the ALBERT model
        albert_model = TFAlbertModel.from_pretrained('albert-base-v2')

        # Get the main layer
        albert = albert_model.layers[0]

        # Build your model input
        input_ids = Input(shape=(max_length,), name='input_ids', dtype='int32')
        inputs = {'input_ids': input_ids}

        # Load the Transformers ALBERT model as a layer in a Keras model
        albert_outputs = albert(inputs)[0]
        albert_model_output = tf.squeeze(albert_outputs[:, -1:, :], axis=1)
        dropout = Dropout(0.1, name='pooled_output')
        pooled_output = dropout(albert_model_output, training=False)

        # Build the model output
        num_labels = len(data_train.Sentiment_label.value_counts())
        Sentiments = Dense(units=num_labels, kernel_initializer=tf.keras.initializers.TruncatedNormal(stddev=0.02), name='Sentiment')(pooled_output)
        outputs = {'Sentiment': Sentiments}

        # Combine it all in a model object
        model = Model(inputs=inputs, outputs=outputs, name='ALBERT_MultiClass')

        # Compile the model
        optimizer = Adam(learning_rate=5e-05, epsilon=1e-08, clipnorm=1.0)
        loss = {'Sentiment': CategoricalCrossentropy(from_logits=True)}
        metrics = {'Sentiment': 'accuracy'}
        model.compile(optimizer=optimizer, loss=loss, metrics=metrics)

        return model


# Print the model summary
model_albert = create_model(max_length)
model_albert.summary()
```

```python
# Ready output data for the model
y_train = to_categorical(data_train['Sentiment'])

# Tokenize the input (takes some time)
tokenizer = AlbertTokenizer.from_pretrained('albert-base-v2')
x_train = tokenizer(
    text=data_train['Phrase'].to_list(),
    add_special_tokens=True,
    max_length=max_length,
    padding=True,
    truncation=True,
    return_tensors='tf',
    return_token_type_ids=False,
    return_attention_mask=False,
    verbose=True)

y_val = to_categorical(data_val['Sentiment'])

x_val = tokenizer(
    text=data_val['Phrase'].to_list(),
    add_special_tokens=True,
    max_length=max_length,
    padding=True,
    truncation=True,
    return_tensors='tf',
    return_token_type_ids=False,
    return_attention_mask=True,
    verbose=True)

# Fit the model under the strategy scope
with strategy.scope():
    history = model_albert.fit(
        x={'input_ids': x_train['input_ids']},
        y={'Sentiment': y_train},
        validation_data=({'input_ids': x_val['input_ids']}, {'Sentiment': y_val}),
        batch_size=64,
        epochs=2,
        verbose=1)
```

```python
import tensorflow as tf
from transformers import ElectraTokenizer, TFElectraModel
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.initializers import TruncatedNormal
from tensorflow.keras.utils import to_categorical

# Define your model creation function
def create_model(max_length):
    with strategy.scope():
        # Load the ELECTRA model
        electra_model = TFElectraModel.from_pretrained('google/electra-base-discriminator')

        # Get the main layer
        electra = electra_model.layers[0]

        # Build your model input
        input_ids = Input(shape=(max_length,), name='input_ids', dtype='int32')
        inputs = {'input_ids': input_ids}

        # Load the Transformers ELECTRA model as a layer in a Keras model
        electra_outputs = electra(inputs)[0]
        electra_model_output = tf.squeeze(electra_outputs[:, -1:, :], axis=1)
        dropout = Dropout(0.1, name='pooled_output')
        pooled_output = dropout(electra_model_output, training=False)

        # Build the model output
        num_labels = len(data_train.Sentiment_label.value_counts())
        Sentiments = Dense(units=num_labels, kernel_initializer=TruncatedNormal(stddev=0.02), name='Sentiment')(pooled_output)
        outputs = {'Sentiment': Sentiments}

        # Combine it all in a model object
        model = Model(inputs=inputs, outputs=outputs, name='ELECTRA_MultiClass')

        # Compile the model
        optimizer = Adam(learning_rate=5e-05, epsilon=1e-08, clipnorm=1.0)
        loss = {'Sentiment': CategoricalCrossentropy(from_logits=True)}
        metrics = {'Sentiment': 'accuracy'}
        model.compile(optimizer=optimizer, loss=loss, metrics=metrics)

        return model

# Print the model summary
model_electra = create_model(max_length)
model_electra.summary()
```

```python
# Ready output data for the model
y_train = to_categorical(data_train['Sentiment'])

# Tokenize the input (takes some time)
tokenizer = ElectraTokenizer.from_pretrained('google/electra-base-discriminator')
x_train = tokenizer(
    text=data_train['Phrase'].to_list(),
    add_special_tokens=True,
    max_length=max_length,
    padding=True,
    truncation=True,
    return_tensors='tf',
    return_token_type_ids=False,
    return_attention_mask=False,
    verbose=True)

y_val = to_categorical(data_val['Sentiment'])

x_val = tokenizer(
    text=data_val['Phrase'].to_list(),
    add_special_tokens=True,
    max_length=max_length,
    padding=True,
    truncation=True,
    return_tensors='tf',
    return_token_type_ids=False,
    return_attention_mask=True,
    verbose=True)

# Fit the model under the strategy scope
with strategy.scope():
    history = model_electra.fit(
        x={'input_ids': x_train['input_ids']},
        y={'Sentiment': y_train},
        validation_data=({'input_ids': x_val['input_ids']}, {'Sentiment': y_val}),
        batch_size=64,
        epochs=2,
        verbose=1)
```
